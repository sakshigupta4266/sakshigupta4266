```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
df = pd.read_csv(r'C:\Users\HP\Downloads\archive (2)\heart.csv')
print(df.head(10))
```

```python
print(df.isnull().sum())
```

```python
# dropna (cleaning the data)
data = {
    'age': [63, 37, 41, 56, 57, np.nan, 56, 44, 52],
    'chol': [233, 250, 204, 236, 354, 192, 294, 263, np.nan],
    'sex': [1, 1, 0, 1, 0, 1, np.nan, 1, 1],
    'target': [1, 1, 1, 1, 1, 1, 1, 1, 1]
}
heart_df = pd.DataFrame(data)
all_missing_df = heart_df.dropna(how='all')
print("DataFrame after removing rows where ALL values are missing:")
print(all_missing_df)
```

```python
# crosstab frequency ke table bana mei help karta haan 
pd.crosstab(df.target, df.sex).plot(kind="bar", figsize=(10, 6), color=["red", "yellow"])
plt.title("Heart Disease Frequency for Sex",fontsize="22")
plt.xlabel("0 = No Disease, 1 = Disease",fontsize="18")
plt.ylabel("Amount",fontsize="18")
plt.legend(["Female", "Male"])
plt.xticks(rotation=0)
plt.show()
```

```python
# scatter with the help positive heart disease
# (thalach term of medical data help of the programming)
plt.scatter(df.age[df.target == 1],
            df.thalach[df.target == 1],
            color="red")
```

```python
# scatter with the help of negative heart no disease
# (thalach term of medical data help of the programming)
plt.scatter(df.age[df.target == 0], 
           df.thalach[df.target ==0],
           color = "yellow")
```

```python
plt.scatter(df.age[df.target == 1],
            df.thalach[df.target == 1],
            color="red")

plt.scatter(df.age[df.target == 0], 
           df.thalach[df.target ==0],
           color = "yellow")

plt.title("Heart Disease in Function of Age and Max Heart Rate")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(["Disease","No Disease"])
plt.show()
```

```python
# create the histrogram of the age
df.age.plot.hist(bins=20)
plt.title("Distribution of the Age",fontsize="22")
plt.xlabel("Age",fontsize="18")
plt.show()
```

```python
# Plot a bar chart showing the count of each chest pain type for both target outcomes
sbn.countplot(x='cp', data=df, hue='target')
plt.title('Heart Disease Frequency Per Chest Pain Type',fontsize="22")
plt.xlabel('Chest Pain Type (0-3)',fontsize="18")
plt.ylabel('Count',fontsize="18")
plt.legend(['No Disease', 'Disease'])
plt.xticks(rotation=0)
plt.show()
```
