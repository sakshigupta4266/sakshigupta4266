```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv(r"C:\Users\HP\Downloads\archive (1)\netflix_titles.csv")
print(df.head())
```

```python
# check missing data
import pandas as pd
print(df.isnull().sum()) 
```

```python
# fill or drop missing data 
import pandas as pd
df = df.dropna(subset=['release_year'])
```

```python
# content types (Movies and  TV shows) count plan
import pandas as pd
type_counts = df['type'].value_counts()
type_counts.plot(kind="bar" , color=['blue','yellow']) 
plt.title("Content Type On NETFLIX",fontsize="22")
plt.xlabel("Type",fontsize="18")
plt.ylabel("Count",fontsize="18")
plt.show()
```

```python
# Top 10 Countries with Most Content
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', color='skyblue')
plt.title("Top 10 Countries with Most Netflix Content",fontsize="22",color="blue")
plt.xlabel("Number of Titles",fontsize="18")
plt.ylabel("Country",fontsize="18")
plt.show()
```

```python
#content Added Over the Years
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
yearly_counts = df['year_added'].value_counts().sort_index()

yearly_counts.plot(kind='line',marker='o', color='purple')
plt.title("Content Added to Netflix Over the Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.show()
```

```python
# Movie Duration Distribution
plt.figure(figsize=(10,6))
plt.hist(df['release_year'], bins=20, color='skyblue')
plt.title("Distribution of Movie Durations",fontsize='22')
plt.xlabel("Duration (mintues)",fontsize='18')
plt.ylabel("Number of Movies",fontsize='18')
plt.grid(axis='y',linestyle='--')
plt.show()
```
